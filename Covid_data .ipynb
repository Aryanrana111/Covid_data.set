```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid", font_scale=1.2)
palette = sns.color_palette("Set1")

plt.figure(figsize=(7,5))
sns.countplot(data=df, x='SEX', palette=palette)
plt.title('Gender Distribution', fontsize=16, weight='bold')
plt.xlabel('Sex (1=Male, 2=Female)')
plt.ylabel('Number of Patients')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,4))
sns.countplot(data = df , x = "PATIENT_TYPE")
plt.title("Patient Type Distribution")
plt.xlabel("1= Outpatient, 2 = Hospitalized")
plt.ylabel("count")
plt.show()
```

```python
sns.countplot(data = df ,x = "DIED")
plt.title("Death Status Count")
plt.show()
```
